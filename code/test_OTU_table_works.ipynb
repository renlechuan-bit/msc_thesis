```python
import pandas as pd
import numpy as np
import os

# === Êñá‰ª∂Ë∑ØÂæÑ ===
unifrac_file = "/home/lechuanr/punim2504/metadata_and_rank_tables_2025-06-11/L2-UniFrac-Paper/mydata/L1_extend_out_unifrac.tsv"

# === Ëá™Âä®ËØÜÂà´Ê†ºÂºèÂπ∂ËØªÂèñ ===
with open(unifrac_file) as f:
    first_line = f.readline()
    second_line = f.readline()

is_matrix_format = first_line.startswith('\t') or len(first_line.strip().split('\t')) == len(second_line.strip().split('\t'))

if is_matrix_format:
    print("üîç Ê£ÄÊµãÂà∞ÊòØÂØπÁß∞ÊñπÈòµÊ†ºÂºè")
    df = pd.read_csv(unifrac_file, sep="\t", index_col=0)
    num_samples = df.shape[0]

    # ÊèêÂèñ‰∏ä‰∏âËßíÔºà‰∏çÂê´ÂØπËßíÁ∫øÔºâÊâÄÊúâÂÄº
    triu_values = df.values[np.triu_indices(num_samples, k=1)]
    actual_pairwise_count = np.sum(~np.isnan(triu_values))

    print(f"üìä ÂÆûÈôÖ pairwise Ë∑ùÁ¶ªÊï∞ÔºàÈùûÁ©∫ÂÄºÔºâ: {actual_pairwise_count}")
    expected = num_samples * (num_samples - 1) // 2
    print(f"üìê ÁêÜËÆ∫Â∫îÊúâ pairwise Êï∞: {expected}")

    if actual_pairwise_count == expected:
        print("‚úÖ Ë∑ùÁ¶ªÁü©ÈòµÂÆåÊï¥")
    else:
        print("‚ö†Ô∏è Ë∑ùÁ¶ªÁü©ÈòµÂèØËÉΩÂ≠òÂú®Áº∫Â§±ÂÄº")
else:
    print("üîç Ê£ÄÊµãÂà∞ÊòØÈïøÊ†ºÂºè")
    df = pd.read_csv(unifrac_file, sep="\t")
    if {"Sample1", "Sample2", "Distance"}.issubset(df.columns):
        count = df.dropna(subset=["Distance"]).shape[0]
        print(f"üìä ÂÆûÈôÖ pairwise Ë∑ùÁ¶ªÊï∞: {count}")
    else:
        print("‚ö†Ô∏è Êú™ÊâæÂà∞ 'Sample1', 'Sample2', 'Distance' ‰∏âÂàóÔºåÊ£ÄÊü•Ê†ºÂºèÊòØÂê¶Ê≠£Á°Æ")
```
